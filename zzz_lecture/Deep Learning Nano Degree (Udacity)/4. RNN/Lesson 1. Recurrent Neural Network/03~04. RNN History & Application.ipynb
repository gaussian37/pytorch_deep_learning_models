80년대 이후, 첫 번째 Neural Network 붐이 있었을 때 겪은 문제점 중 하나는 Feedforward 네트워크는 Temporal Dependency (시간 의존성)를 적용할 수 없다는 것이었습니다.<br>
실제 세상에 사용되는 많은 데이터들은 시간의 변화에 따른 값을 가집니다. 따라서 음성, 비디오 등의 natural signal 은 시간에 따라 값이 변하고 특성 또한 변하여 Feedforward Network만을 사용하기에는 한계가 있었습니다.

![2](http://postfiles8.naver.net/MjAxODAxMjFfMTgx/MDAxNTE2NTAwMjgzMjA2.aWrx7SYeBB8Vq7Ey61IptkvGP3Q8lxp7NvzUbQeN9H4g.GVTqjJNCe2Oc3LzPxh04B25O4GQvZYs4ma-U412ccbwg.PNG.infoefficien/image_5738859351516500265451.png?type=w773)

실제 Biological Neural Network는 Recurrent 성질을 가지므로 시간에 따른 영향을 받는 데이터는 Recurrent 성질을 이용하여 과거/현재 데이터를 입력 받는 것이 좀 더 타당해 보입니다. 

![3](http://postfiles8.naver.net/MjAxODAxMjFfNzMg/MDAxNTE2NTAwNTE1MDIy.tj087rExgn1cAPIM7k8akIpCJA3rgmbChzoQyJNbZzog.UQCGwBSoYcLsi1PVqHEG0cpSp-WfwhqHcNNhXwQuWIog.PNG.infoefficien/image_5294050411516500501620.png?type=w773)

첫 번째로 Neural Network에 memory 구조를 더한 것은 TDNN(Time Delay Neural Network)입니다. TDNN은 이전 step의 input이 현재 step의 input으로도 사용되는 구조를 가집니다.<br>
TDNN은 과거의 input을 현재도 사용할 수 있다는 장점을 가지지만, 정해진 time step에 한해서만 과거 데이터를 사용할 수 있는 단점이 있습니다.(슬라이딩 윈도우 형식으로 과거 데이터 참조)

![4](http://postfiles2.naver.net/MjAxODAxMjFfMTA5/MDAxNTE2NTAwOTcwMDA3.y2Yfx2Mfblk_6WInE94YREPifdTkgseCIQo8a9cI0R4g.qWhhYhqHVHxshlcrJ28B_ZxdxzLCHL8fgwfyYR6tDoQg.PNG.infoefficien/image_7039211581516500953518.png?type=w773)

TDNN을 좀 더 개선한 가장 간단한 RNN 구조를 가지는 ELMAN 네트워크를 추후에 알아보도록 하겠습니다. 

![5](http://postfiles1.naver.net/MjAxODAxMjFfMjM0/MDAxNTE2NTAxMTE4NDI0.y26auYjir0c7DVrXGyt9uSzNeoSdHohng3a6BeMGIA0g.LmJ4Rk01r73Zpa9nHH7PyWn5JP_AMA2CNd_XrWxR9Sog.PNG.infoefficien/image_3311500151516501101823.png?type=w773)

90년대에는 RNN 유사 Network에서 가지는 공통 문제는 Vanishing Gradient 였습니다. Gradient값 전파 시 Layer를 거칠 수록 Gradient 값이 기하급수적으로 작아졌습니다. 따라서 RNN에서 Vanishing Gradient 문제를 해결하는 방법 또한 추후에 자세히 알아보도록 하겠습니다.

![6](http://postfiles2.naver.net/MjAxODAxMjFfMjgg/MDAxNTE2NTAxMzcxODcz.MYSjnxrpPOTT2PxLAQoD1PgzvYr3by5DeX3npOqHVVsg.Iyx35Bt8DSKKAJnk7_idu75-eGA-sdZzczcutF86kAwg.PNG.infoefficien/image_4983542001516501307411.png?type=w773)

이후 90년대 중반 LSTM(Long Short Term Memory cell)이 앞에서 제기 되었던 문제들을 해결하기 위해 등장하였습니다. LSTM에서는 state variable이라고 부르는 signal이 고정이 되어 Network에서 이후 time에 적절히 사용되어 Netowkr에 time dependency를 적용할 수 있습니다.<br>
LSTM은 추후에 자세히 공부할 예정입니다. LSTM을 다 공부하고 나면 LSTM의 파생형힌 GRU(Gated Recurrent Network)를 공부해 보겠습니다.
