FFNN에서는 output은 현재 input과 weight에 의해서 결정됩니다. 

![2](http://postfiles13.naver.net/MjAxODAxMjRfMTM2/MDAxNTE2Nzk5MzYzNzc1.r7Dk4hjYgxb2mknfA8FJGHt94L092ay240bdKk1bXcEg.pU-KWwr7h3M8Klh5g5PJ8NUftrWc1xAznHmPXb6pBGMg.PNG.infoefficien/13._RNN%28b%29.mp4_000037076.png?type=w773)

RNN에서는 time = t 일 때,  현재 input인 $x_{t}$뿐만 아니라 $x_{t-1}, x_{t-2}, ..., x_{t-N}$ 등도 input에 영향을 줍니다. RNN 내부를 좀 더 시각화 해보겠습니다.

![3](http://postfiles9.naver.net/MjAxODAxMjRfMjky/MDAxNTE2ODAwOTA5ODc3.i96bMNLae-TpWujkySjK0AwkKM1zyJktz5Su5PorsbYg.UCNhFNlFt5ZSYTySrZHPHr-xQ71aZCHzBbsoGDiV-BMg.PNG.infoefficien/13._RNN%28b%29.mp4_000134008.png?type=w773)

RNN에서는 input $\overline{x}$와 ouput $\overline{y}$ 그리고 hidden layer $\overline{s}$를 가집니다. 여기서 $\overline{s}$는 state를 나타내고 이것은 memory를 가집니다. $W_{x}$는 input과 state layer를 연결하는 weight matrix를 나타냅니다. $W_{y}는 state와 output을 연결하는 weight matrix 입니다. $W_{s}$는 이전 time step과 다음 time step의 state를 연결해주는 weight matrix 입니다. 위와 같은 형태로 표현한 모델을 Folded model 이라고 합니다. 
Folded model 에서는 입력된 input은 여러번 계속 반복되어 입력되게 되고 같은 연산을 반복하게 됩니다. 

![4](http://postfiles1.naver.net/MjAxODAxMjRfMjQ0/MDAxNTE2ODAwNjE3MTk4.hZn34lQZDj0RX3yGY6EoNR4OQy2hlyh1pQ9XrvvVnRog.tbtw4nlAxtREScpoad44QclvcQClwpJ4Epnoo1HMenYg.PNG.infoefficien/13._RNN%28b%29.mp4_000115803.png?type=w773)

Unfolded model에 대하여 알아봅시다. Unfolded model은 folded model를 전개하여 표현한 방식 입니다. 예를 들어 현재 step이 t+2라고 할때, input은 $x_{t+2}$이고 output은 $y_{t+2}$ 그리고 위의 노란색 박스는 이전 input에 해당하여 현재 step t+2에 영향을 주게 됩니다.

![5](http://postfiles2.naver.net/MjAxODAxMjRfMjE5/MDAxNTE2ODAwOTg4NzUw.Rabg4gA_Cdvbz-mi7q9pN7YQp0xNZgfvav6neaMmJa8g.qFbymOnAQe_ou1fJC_bmiLfJxwVXNjp8Ra5b-V_s6CIg.PNG.infoefficien/13._RNN%28b%29.mp4_000147803.png?type=w773)

먼저 FFNN에서 hidden layer인 h를 얻기 위해서는 input과 weight를 곱한 후 activation을 적용하였습니다. 

![6](http://postfiles12.naver.net/MjAxODAxMjRfMjAy/MDAxNTE2ODAxMDgyMTA4.1-Ij7Jx_r8xbYjHkQxXS5YyyF7ovTFD3lokAiepL_-Ug.gb_4Gr6MVYC6MTvVZ4RxtUI_GbiYG5O0i3w9UMxSwNYg.PNG.infoefficien/13._RNN%28b%29.mp4_000192666.png?type=w773)

RNN 에서도 hidden layer인 $\overcome{s}$를 얻기 위해서 activation function을 적용해야 합니다.
FFNN에서는 $h = \phi(\overline{x} \cdotp W_{x})$ 이지만 RNN에서는 $h = S_{t} = \phi(\overline{x} \cdotp W_{x} + S_{t-1} \cdotp W_{s})$ 가 됩니다. 여기서 $S_{t-1}$은 이전의 activation 값입니다. 
마지막으로 output을 생성할 때에는 $S_{t} \cdotp W_{y} $의 곱 또는 softmax를 적용한 $\sigma(\overline{S_{t}} \cdotp W_{y})$가 됩니다.

![7](http://postfiles9.naver.net/MjAxODAxMjRfOTQg/MDAxNTE2ODAyOTY3OTg5.mQ7nLjr3YMrfCI6yRhcKCgqKGAU7ZUyDgvAQCuOhjigg.YBtViPnCx6KZ-546qNlhdaB1Hqma68dDgNvb1FbT1EIg.PNG.infoefficien/13._RNN%28b%29.mp4_000199299.png?type=w773)

RNN은 매 번의 <span class="mark">time step 마다 동일한 parameter를 공유</span> 합니다. 
RNN은 memory 구조로 인하여 FFNN보다 다소 복잡한 구조를 가지고 있습니다. 하지만 동일한 parameter를 재사용 하기 때문에 parameter의 수는 그다지 크지 않습니다. 
위와 같은 Unfolded model과 같은 방식으로 RNN을 표현하는 것이 일반적이지만 사용할 Network architecture에 따라서 구조를 변형 할 수 있습니다. 따라서 Unfolded model 방식으로 표현할 때에는 <span class="mark">얼마나 많은 input과 output이 필요한 지 결정</span>해야 합니다. 

![8](http://postfiles9.naver.net/MjAxODAxMjRfMjMy/MDAxNTE2ODA0NTQ4NTc1.gGDPdN-Dhc1kX_EoARPWFCYjLHrUhAbneocPQ0dm9Qsg.fcTSGyiaqoXil1Z3MUmgm-y21E2XDgFy0clZDiBwZ9Mg.PNG.infoefficien/13._RNN%28b%29.mp4_000234377.png?type=w773)

예를 들어 위와 같이 single level의 RNN을 구성하여 output으로 vector $\overline{y}$를 만들 수도 있지만,

정리하면,<br>
FFNN에서는 출력은 현재 입력 및 가중치에 관한 함수입니다. 이것은 다음 방정식을 사용하여 쉽게 나타낼 수 있습니다.

$$ \overline{y}_{t} = F(\overline{x}_{t}, W) $$

RNN에서 우리의 시간 t에서의 출력은 현재 입력과 가중치뿐만 아니라 이전 입력에도 의존합니다. 이 경우 시간 t에서의 출력은 다음과 같이 정의됩니다.

$$ \overline{y}_{t} = F(\overline{x}_{t}, \overline{x}_{t-1}, \overline{x}_{t-2}, ..., \overline{x}_{t-t_{0}}, W)$$

| 기호 |      내용   |
|------|      ------|
|$\overline{x}$|&nbsp;&nbsp;&nbsp;&nbsp;input vector|
|$\overline{y}$|&nbsp;&nbsp;&nbsp;&nbsp;output vector|
|$W_{s}$|&nbsp;&nbsp;&nbsp;&nbsp;input과 state를 연결하는 weight matrix|
|$W_{y}$|&nbsp;&nbsp;&nbsp;&nbsp;state와 output을 연결하는 weight matrix|
|$W_{s}$|&nbsp;&nbsp;&nbsp;&nbsp;이전 step의 state와 현재 state를 연결하는 weight matrix|
